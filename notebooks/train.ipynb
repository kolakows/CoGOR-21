```python
import pandas as pd

final_train = pd.read_csv('final_train.csv')

cols_to_drop = ['Unnamed: 0', 'void()']
final_train = final_train.drop(cols_to_drop, axis = 1)

removed_subjects = [28, 9]
final_train = final_train.query('subject not in @removed_subjects')
final_train = final_train.drop(['subject'], axis = 1)
final_train = final_train.drop_duplicates()

cols_to_keep = final_train.columns
```

```python
train_x, train_y = final_train[final_train.columns.difference(['Activity'])], final_train['Activity']
```

```python
train_x.head(3)
```

```python
train_y.head(3)
```

```python
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb

model = xgb.XGBClassifier()
model.fit(train_x, train_y)

y_pred = model.predict(train_x)
```

```python
train_y.values
```

```python
accuracy_score(train_y.values, y_pred)
```

```python
import pickle

fi_model = './xgboost_model.pkl'
fi_cols = './columns_model.pkl'
cols = cols_to_keep.difference(['Activity'])

with open(fi_cols, 'wb') as f:
  pickle.dump(cols, f)

with open(fi_model, 'wb') as f:
  pickle.dump(model, f)
```
