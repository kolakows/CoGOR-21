```python
import pandas as pd

final_train = pd.read_csv('final_train.csv')
final_train = final_train.drop(['Unnamed: 0', 'void()'], axis = 1)
removed_subjects = [28, 9]
final_train = final_train.query('subject not in @removed_subjects')
final_train.head(1)
```

```python
final_train = final_train.drop('subject', axis = 1)
```

```python
len(final_train.subject.unique())
```

```python
target = 'Activity'

X = final_train[final_train.columns.difference([target])]
Y = final_train[target]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb

kfold = StratifiedKFold(n_splits=5, random_state=123, shuffle = True)
model = xgb.XGBClassifier()
results = cross_val_score(model, X, Y, cv=kfold)
```
